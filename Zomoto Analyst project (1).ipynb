## Step -1 - Import Library for visualization and Action perform

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Step-2 - Create dataframe

```python
df = pd.read_csv("Zomato data .csv")
```

```python
df.head()
```

## Convert the data type of column -Rate

```python
def handleRate(value):
    value = str(value).split('/')
    value = value[0];
    return float(value)

df['rate'] = df['rate'].apply(handleRate)
print(df.head())
```

## Type of resturant

```python
df.head()
```

```python
sns.countplot(x=df['listed_in(type)'])
plt.xlabel("Type of Resturant")
plt.show()
```

```python
data = df.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes': data})
plt.plot(result, c='Green', marker='o')
plt.xlabel('Type of Resturant', c='Red', size=20)
plt.ylabel('Votes', c='Red' , size=20)
plt.show()
```

```python
df.head()
```

```python
plt.hist(df['rate'], bins=5)
plt.title("Rating Distribution")
plt.show()
```

## Average order spending on copules

```python
copule_data = df['approx_cost(for two people)']
sns.countplot(x=copule_data)
plt.show()
```
